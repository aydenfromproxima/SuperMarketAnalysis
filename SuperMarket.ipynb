```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(x=df['Gender'])
plt.title("Total Number Of Male and Female")
```

```python
sns.distplot(df['Rating'])
plt.grid()
```

```python
df["Branch"].value_counts()
```

```python
sns.countplot(x=df['Branch'],palette='tab20')
plt.title("Count Vs Branch")
```

```python
payment_counts = df['Payment'].value_counts()
print(payment_counts)
sns.countplot(data=df, x='Payment')
plt.title("Popular Payment Methods")
plt.show()
```

```python
payment_by_branch = df.groupby(['Branch', 'Payment']).size().unstack()
payment_by_branch.plot(kind='bar', stacked=True)
plt.title("Payment Method Distribution by Branch")
plt.ylabel("Number of Transactions")
plt.show()
```

```python
payment_by_branch = df.groupby(['Branch', 'Payment']).size().unstack()
payment_by_branch.plot(kind='bar', stacked=True)
plt.title("Payment Method Distribution by Branch")
plt.ylabel("Number of Transactions")
plt.show()
```

```python
sns.boxplot(data=df, x='Branch', y='Rating')
plt.title("Customer Ratings by Branch")
plt.show()
```

```python
profit_by_branch = df.groupby('Branch')['gross income'].sum()
print(profit_by_branch)

profit_by_branch.plot(kind='bar', title="Total Gross Income by Branch")
plt.ylabel("Gross Income")
plt.show()
```

```python
sns.boxplot(data=df, x='Gender', y='gross income')
plt.title("Gross Income by Gender")
plt.show()
# mean income by gender
print(df.groupby('Gender')['gross income'].mean())
```

```python
# Convert date if not already
df['Date'] = pd.to_datetime(df['Date'])

# Daily income trend
daily_income = df.groupby('Date')['gross income'].sum()

daily_income.plot(title="Daily Gross Income Trend")
plt.ylabel("Gross Income")
plt.xlabel("Date")
plt.show()
```
